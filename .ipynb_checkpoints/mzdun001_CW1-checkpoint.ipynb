```python
%matplotlib inline
'''
Takes in data X and targets y. Creates a figure with multiple subplots (16) 4 by 4 plots comparing each of the features
- diagonal subplots show a histogram on that feature. 
'''
def myplotGrid(X,y):
    # create a figure with multiple subplots
    fig = plt.figure(figsize=(12,12))
    features = ['Sepal length (cm)', 'Sepal width (cm)', 'Petal length (cm)', 'Petal width (cm)']
    classes = ['Setosa','Versicolour','Virginica']
    cdict = {0:'mediumaquamarine',1:'lightsalmon',2:'lightsteelblue'}
    index = 0
    for i in range(len(features)):
        
        for j in range(len(features)):
           
            index += 1
            # 4 rows and 4 columns, index is the number of the current subplot
            ax = fig.add_subplot(4,4,index)
            
            # for the feature legend (check for j and i so it's only on the right and at the bottom)
            if(j==0):
                plt.ylabel(features[i])
            if(i==3):
                plt.xlabel(features[j])
                
            if(i==j):
                ax.hist(X[:,i], color='lightskyblue')
            else:
                
                #ax.scatter(X[:, j], X[:, i], c=y)
                
                #had a problem with adding a legend so did it with help of this: 
                #https://stackoverflow.com/questions/47006268/matplotlib-scatter-plot-with-color-label-and-legend-specified-by-c-option
                
                # j and i are columns in data set (features, e.g. sepal length)
                Xj = X[:,j]
                Xi = X[:,i]
            
                # scatter each class separately (because of the legend)
                for k in np.unique(y):
                    target = np.where(k == y)
                    ax.scatter(Xj[target], Xi[target], c=cdict[k], s= 20, label=classes[k])
                    # check for i and j to display legend (right upper corner picture)
                    if(i==0 and j==3):
                        ax.legend()
```

```python
##################### HELPER FUNCTIONS ########################

def euclideanDistance(in1,in2):
    dist = numpy.linalg.norm(in1-in2)
    return dist


def manhattanDistance(in1, in2):
    dist = 0.0
    for i in range(len(in1)):
        dist += abs(in1[i] - in2[i])
    return dist

# the following function written with help of tutorial : 
#https://machinelearningmastery.com/tutorial-to-implement-k-nearest-neighbors-in-python-from-scratch/   
'''
getNeighbours takes in a test data point (x_), training data (X), number of neighbours (n) and distance function (dist_f);
It loops through all training data and calls the distance function to calculate the distance between the test data point
and all data points in the training set. It saves both the index of the point in training set and the distance.
It sorts that lists by distances and returns only the first n indices of data points in training set.
'''
def getNeighbours(x_,X,n,dist_f): # 
    distances = []
    for i in range(len(X)):
        dist = dist_f(x_,X[i])
        distances.append((i,dist))
    distances.sort(key=lambda x: x[1]) #sort by distances (x[1] for x in the distances list)
    neighbors = [item[0] for item in distances] #only need indices of neigbours not the actual distances
    neighbors = neighbors[:n] #return first n items from the list (number of neighbours)

    return neighbors # indices of n-nearest neighbours in training data

'''
assignLabel takes in the number of labels of of n-nearest neighbours of test data point. 
It checks if the length of the list is equal to 0 or 1 (if there is only one element in the list, returns only 
that one element). 
If the length is greater than 1 then it loops through the list counting the number of each labels in the list 
and returns the label that occurs the most in the list.
'''
def assignLabel(nLabels):
    if(len(nLabels) == 0):
        return ("Empty list") 
    if(len(nLabels) < 2):
        return nLabels[0] #i.e. if there is only one element in the list of labels
    else: 
        maxCount = 0 
        index = 0 # so we can keep track of the index of assigned label
        # the two for loops: for each element in the list, count the number of the elements the same as this element
        for i in range(0,len(nLabels)): # (looping through indices of nLabels)
            count = 0
            for j in nLabels: #(looping through elements of nLabels)
                if(nLabels[i]==j):
                    count += 1
            if(count > maxCount): #if count of the current element is higher than maximum
                maxCount = count # count is the new maximum
                index = i # save the index of that label
    return nLabels[i] # label assigned to test point x_ 
```

```python
import numpy

'''
myKNN function takes in training data X, training labels y, test data X_, number of neighbours nn and distance function 
dist_function. It loops through all test data points, calls for getNeighbours to find neighbours of each point,
creates a list of labels of each neigbour and calls assignLabel to assign label based on a majority vote.
The function returns a list of predictions (list of predicted labels).
'''
def myKNN(X,y,X_,nn,dist_function):
    
    predictions = []
    true_labels = []
    for i in range(len(X_)): #for all test points
        # knn classifier
        x_=X_[i] # test point x_
        #y_=y[i] # true label for x_
        #true_labels.append(y_)
        #print(y_)

        # get neighbours of x_ in training data 
        neighbours = getNeighbours(x_,X,nn,dist_function)
        # assignLabel to x_ based on neighbours
        pred_labels = y[neighbours]
        predY = assignLabel(pred_labels)
        predictions.append(predY)
        
    #acc_ = accuracy(predictions,true_labels)
    #return ((predictions, acc_))
    return predictions
        
np.random.seed(mySeed)
indices= np.random.permutation(X.shape[0]) 
bins=np.array_split(indices,2) 
foldTrain=bins[0]
foldTest=bins[1]

# Train / test split for data and labels
X_train, y_train, X_test, y_test = X[foldTrain], y[foldTrain], X[foldTest], y[foldTest]
#print(y_train)

# distance function and number of neighbours variables
dist_f = euclideanDistance
nn = 10

# call function
y_ = myKNN(X_train,y_train,X_test,nn,dist_f)
#y_ = myKNN(X,y,nn,dist_f,mySeed)
print(y_)
print(y_test)

# function to calculate accuracy (compares predictions with true labels)
def accuracy(predictions,y_test):
    correct=0;
    for i in range(len(y_test)):
        if(predictions[i] == y_test[i]):
            correct += 1
    return (correct/float(len(y_test)))*100

print("ACCURACY :",accuracy(y_,y_test),"% for distance function: ", dist_f.__name__, ", nn: ", nn)           
```

## 3. Nested Cross-validation using your implementation of KNN

In the cell below, develop your own code for performing 5-fold nested cross-validation along with your implemenation of k-NN above. Within this you should write code that evaluates classifier performance. You must write your own code -- the scikit-learn module may only be used for verification purposes. 

Your code for nested cross-validation should invoke your kNN function (see above). You cross validation function should be invoked similary to:

    accuracies_fold, best_parameters_fold = myNestedCrossVal(X,y,5,list(range(1,11)),['euclidean','manhattan'],mySeed)
    
where X is your data matrix (containing all samples and features for each sample), 5 is the number of folds, y are your known output labels, ``list(range(1,11)`` evaluates the neighbour parameter from 1 to 10, and ``['euclidean','manhattan']`` evaluates the two distances on the validation sets.  mySeed is simply a random seed to enable us to replicate your results. The outputs could be a list of accuracy values, one per fold, and a list of the corresponding parameter tuples ``(distance, k)`` used to calculate these.

**Notes:** 
- you should perform nested cross-validation on both your original data X, as well as the data pertrubed by noise as shown in the cells above (XN)
- you should implement/validate **at least** two distance functions
- you should evaluate number of neighbours from 1 to 10
- your function should return a list of accuracies per fold, and a list of the corresponding parameters

  - the accuracy per distinct set of parameters on the validation set

  - the confusion matrix per fold (on the testing set)

```python
'''
myConfMat takes in true labels (y_test), predictions (y_pred) and number of classes (classno).
It creates a matrix of zeros classno x classno.
Then it loops through all results and updates the confusion matrix, and outputs the updated matrix.
'''
def myConfMat(y_test,y_pred,classno):
    C= np.zeros((classno,classno)) # initialize the confusion matrix to zeros
    #loop through all results and update the confusion matrix
    for i in range(len(y_test)):
        C[y_test[i]][y_pred[i]] += 1
            
    return C
```

```python
'''
myNestedCrossVal takes in data X, targets y, number of folds foldK, number of neighbours nns, 
distance functions dists, and mySeed.

It splits all data into bins based on number of folds. 
Then loops through the bins, and takes bins[i] for testing, following bin for valdiation and the rest for training. 

In each fold, loops through distances and numbers of neigbours, calls myKNN function to produce a list of predictions and
calls accuracy function to calculate accuracy of these predictions. It saves the best accuracy as bestAccuracy and saves
the parameters (distance function and number of neigbours) that were used to achieve that accuracy.

Then it combines training and validation data and runs myKNN again with the best parameters on test set, 
calculates accuracy of predictions on test set and finally creates a confusion matrix on test set as well. 

The function returns two outputs:
1. (test_acc_fold,best_parameters_fold): List of test accuracies (per fold) and best parameters (per fold)
2. confusion_matrix_fold: confusion matrix (per fold);
'''

def myNestedCrossVal(X,y,foldK,nns,dists,mySeed):
    np.random.seed(mySeed)
    #val_accuracies_fold=[]
    #parameters_fold = []
    
    test_acc_fold = []
    best_parameters_fold = []
    confusion_matrix_fold = []
    
    indices = np.random.permutation(list(range(X.shape[0])))
    bins = np.array_split(indices, foldK)
    
    #loop through folds
    for i in range(0,foldK):
        foldTrain=[] # list to save current indices for training
        foldTest=[]  # list to save current indices for testing
        foldVal=[]    # list to save current indices for validation

        #loop through all bins, take bin i for testing, the next bin for validation, and the rest for training
        foldTest = bins[i] #bin i for testing
        
        # bin i+1 for validation (if we're at the end of the list, validation starts from the beginning of the list)
        if(i < foldK-1):
            val = i+1
            foldVal = bins[val]
        else:
            val = -(i+1)
            foldVal = bins[val]
            #reset to positive index so can check for it later with train
            val = val-val

        #the rest for training
        for j in range(0,len(bins)):
            #everything that's not test or val
            if(j != i):
                if(j != val):
                    #print("J",j)
                    foldTrain.extend(bins[j])
                    
        assert not np.intersect1d(foldTest,foldVal)
        assert not np.intersect1d(foldTrain,foldTest)
        assert not np.intersect1d(foldTrain,foldVal)
    
                    
        X_partial_train, X_val, y_partial_train, y_val = np.array(X[foldTrain]), X[foldVal], y[foldTrain], y[foldVal]
        
        #print("XTRAIN",X_partial_train,"XVAL", X_val,"YTR", y_partial_train,"YVAL", y_val)

        bestDistance='' #save the best distance metric here
        bestNN=-1 #save the best number of neighbours here
        bestAccuracy=-10 #save the best attained accuracy here (in terms of validation)

        acc_list = []
        #par_list = []

        for i in dists:
            for j in nns:
                y_ = myKNN(X_partial_train,y_partial_train,X_val,j,i)
                acc = accuracy(y_,y_val)
                
                acc_list.append(acc)
                #par_list.append(((i.__name__,j)))
                
                if(acc > bestAccuracy):
                    bestAccuracy = acc
                    bestDistance = i
                    bestNN = j


        print('** End of val for this fold,', 'val accuracies: ', acc_list, ' best NN: ', bestNN, ', best dist:', bestDistance.__name__)


        #evaluate on test data:
        #extend training set by including the validation set
        X_train = np.concatenate((X[foldTrain], X[foldVal]))
        X_test = X[foldTest]
        y_train = np.concatenate((y[foldTrain], y[foldVal]))
        y_test = y[foldTest]

        #train k-NN classifier on new training set and test on test set
        y_ = myKNN(X_train,y_train,X_test,bestNN,bestDistance)
        #get performance on fold, save result in accuracy_fold array
        test_acc = accuracy(y_,y_test)
        test_acc_fold.append(test_acc)
        best_parameters = ((bestDistance.__name__,bestNN))
        best_parameters_fold.append(best_parameters)
        
        #print("PREDICTIONS" , y_, "TRUE LABELS",  y_test)
        #val_accuracies_fold.append(acc_list)
        #parameters_fold.append(par_list)
    
        print('==== Final Cross-val on test on this fold; accuracy = ', test_acc)
        conf = myConfMat(y_test,y_,len(np.unique(y)))
        print("+ Confusion matrix on the test set:", "\n",conf)
        confusion_matrix_fold.append(conf)
        
    return (test_acc_fold,best_parameters_fold), confusion_matrix_fold;
```

Precision and recall functions

```python
print(conf[1])
'''
True positive: diagonal position
False positive: sum of column x (without main diagonal)
False negative: sum of row x (without main diagonal)

Precision = TP/(TP+FP)
Recall = TP/(TP+FN)
'''

def precision_recall(confusion_matrix, label):
    TP = confusion_matrix[label][label]
    FP = sum(confusion_matrix[:,label]) - TP
    FN = sum(confusion_matrix[label,:]) - TP
    
    precision = TP/(TP+FP)
    recall = TP/(TP+FN)
    return precision, recall
    
precision, recall = precision_recall(conf[1],1)
print(precision)
print(recall)
```
